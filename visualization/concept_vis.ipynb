# TNCSI

```python
from local_deprecated.metrics.metric import _plot_s2citaions
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
# 生成一个呈现右偏正态分布的随机数据
data = np.random.randn(1000) * 2 + 5  # 均值为5，标准差为2的正态分布随机数
citation, _ = _plot_s2citaions('CLIP', total_num=1000)
citation = np.array(citation)

plt.style.use('seaborn-v0_8')
fig = plt.figure(figsize=(6, 4), dpi=300)  # Increase DPI for high resolution
data = citation
np.mean(data)

params = stats.expon.fit(data)
loc, scale = params
print(loc,scale)
# 生成拟合的指数分布曲线
x = np.linspace(np.min(data), np.max(data), 100)
pdf = stats.expon.pdf(x, loc, scale)
plt.xlabel('Received Citation')
plt.ylabel('Frequency')

plt.hist(data, bins=1000, density=True, color='darkorange', edgecolor='darkorange',alpha=1.0, label='Raw Data')
plt.plot(x, pdf, 'steelblue' ,label='Fitted Curve')

# Find the index where x is just less than 321 for area filling
idx = np.where(x < 400)

# Fill the area under the curve up to x < 321
plt.fill_between(x[idx], pdf[idx], color='lightsteelblue', alpha=0.7, label='Area under the curve = TNCSI')

plt.legend()
# plt.savefig('TNCSI_enhanced.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

# TNCSI Simple Vis

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x from 0 to 1000 with sufficient resolution
x = np.linspace(0, 1000, 100)

# Define the exponential decay function for y
decay_rate = 0.005  # Decay rate; adjust as needed for your curve

# Compute y using the exponential decay function
y = np.exp(-decay_rate * x)

# Normalize y so that the area under the curve (integral) sums to 1
y_normalized = y / np.sum(y)

# Now the rest of your plotting code follows
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(6, 4), dpi=300)

plt.plot(x, y_normalized, color='steelblue', linewidth=2, label='Cumulative Distribution')

# Find the index where x is just less than 321 for area filling
idx = np.where(x < 321)

# Fill the area under the curve up to x < 321
plt.fill_between(x[idx], y_normalized[idx], color='grey', alpha=0.5, label='Area under the curve = TNCSI')

plt.xlabel('Received Citation', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.legend(fontsize=10)
plt.savefig('TNCSI_enhanced.svg', dpi=300, bbox_inches='tight')
plt.show()
```

# RUI

```python

from local_deprecated.metrics.metric import *

# Define the function
def lognorm_function(t, CDR):
    return 10 * get_RAD(t * 12) + 10 * CDR

# Create a range of values for t and CDR
t_values = np.arange(0, 8, 0.1)
CDR_values = np.linspace(0, 2, 100)

# Create a meshgrid for t and CDR values
t_grid, CDR_grid = np.meshgrid(t_values, CDR_values)

# Calculate the function values
lognorm_values = np.vectorize(lognorm_function)(t_grid, CDR_grid)
plt.rcdefaults()

# Plotting the function with emphasis on the line where CDR = 1
fig = plt.figure(figsize=(10, 8), dpi=300)  # Increase DPI for high resolution
ax = fig.add_subplot(111, projection='3d')

# Plotting the surface
surf = ax.plot_surface(t_grid, CDR_grid, lognorm_values, cmap='plasma', alpha=0.8)

# Emphasizing the line where CDR = 1
CDR_one_values = np.vectorize(lognorm_function)(t_values, 1)
ax.plot(t_values, np.ones_like(t_values), CDR_one_values, color='red', linewidth=2, label='CDR = 1')

# Set axis labels and title
ax.set_xlabel('Years After Publication', fontsize=18, labelpad=10)
ax.set_ylabel('CDR', fontsize=18, labelpad=10)
ax.set_zlabel('RUI', fontsize=18, labelpad=10)

# Set tick parameters to increase font size
ax.tick_params(axis='both', which='major', labelsize=16)
ax.tick_params(axis='z', which='major', labelsize=16)

# Add color bar
cbar = fig.colorbar(surf, shrink=0.5, aspect=10, pad=0.1)
cbar.set_label('RUI', rotation=270, labelpad=20, fontsize=18)

ax.legend(loc='upper right', bbox_to_anchor=(1, 0.9))

plt.tight_layout()
plt.savefig('RUI.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from sklearn.linear_model import RANSACRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import numpy as np
from collections import defaultdict
tick_sz = 14
legend_sz = 14
axis_sz = 16
import re
import datetime
def preprocess_field_name(field_name):
    match = re.search(r'\(ti:\"review\" OR ti:\"survey\"\) AND \(ti: \"(.*?)\" OR abs:\"(.*?)\"\)', field_name, re.IGNORECASE)
    if match:
        return match.group(1).lower()
    else:
        return field_name.lower()
def scatter_with_fitted_curve(rows, x_col, y_col, keyword_col, selected_keywords, remove_outliers=False):
    # Preprocess field names
    for row in rows:
        setattr(row, keyword_col, preprocess_field_name(getattr(row, keyword_col)))

    # Filter rows with valid citation counts
    filtered_rows = [row for row in rows if getattr(row, 's2_citation_count', 0) >= 0]
    base_date = datetime.datetime(2020, 1, 1)

    # Extract (x, y, keyword) data
    data = [
        (
            (getattr(row, x_col) - base_date).total_seconds() / (24 * 3600 * 30), 
            getattr(row, y_col), 
            getattr(row, keyword_col)
        )
        for row in filtered_rows
        if getattr(row, x_col) is not None and 
           getattr(row, y_col) is not None and 
           getattr(row, keyword_col) is not None
    ]

    # Filter and categorize data by keyword
    x_data, y_data, keywords = zip(*data)
    filtered_data = [
        (x, y, keyword) 
        for x, y, keyword in zip(x_data, y_data, keywords) 
        if keyword in selected_keywords
    ]

    categorized_data = defaultdict(list)
    for x, y, keyword in filtered_data:
        categorized_data[keyword].append((x, y))
    
    fig, ax = plt.subplots(figsize=(6, 4))
    colors = ['steelblue', '#ED7D31']
    kwds = ['Language Modeling', 'Recommend Systems']
    keyword_to_color = {keyword: colors[i] for i, keyword in enumerate(selected_keywords)}

    i = 0
    for keyword, color in keyword_to_color.items():
        x_vals = np.array([x for (x, y) in categorized_data[keyword]])
        y_vals = np.array([y for (x, y) in categorized_data[keyword]])

        # Optionally remove outliers based on IQR
        if remove_outliers:
            q1, q3 = np.percentile(y_vals, [25, 75])
            iqr = q3 - q1
            lower_bound = q1 - 1.5 * iqr
            upper_bound = q3 + 1.5 * iqr
            mask = (y_vals >= lower_bound) & (y_vals <= upper_bound)
            x_vals, y_vals = x_vals[mask], y_vals[mask]
        if color == 'steelblue':
            edge_clr = 'b'
        else:
            edge_clr = 'r'
        # Scatter plot
        ax.scatter(x_vals, y_vals, alpha=0.7, color=color, label=kwds[i])

        # Use robust regression (RANSAC) for fitting
        ransac = make_pipeline(PolynomialFeatures(degree=1), RANSACRegressor())
        ransac.fit(x_vals.reshape(-1, 1), y_vals)

        # Generate smooth curve
        smooth_x = np.linspace(min(x_vals), max(x_vals), 100)
        smooth_y = ransac.predict(smooth_x.reshape(-1, 1))

        # ax.plot(smooth_x, smooth_y, color=color, lw=2)
        i += 1

    # Configure axis labels and limits
    ax.set_xlabel('Publication Date', fontsize=axis_sz)
    ax.set_ylabel(r'RUI', fontsize=axis_sz)
    ax.set_xlim(0, 57)
    ax.set_xticks(np.arange(0, 57, 12))  # 设定 12, 24, 36, 48 的刻度
    ax.set_xticklabels([f'{int(2020+(i/12))}' for i in np.arange(0, 49, 12)])
    ax.set_ylim(0, 52)
    ax.tick_params(axis='both', which='major', labelsize=tick_sz)  # 主刻度标签
    ax.tick_params(axis='both', which='minor', labelsize=tick_sz)  # 次刻度标签
    # Legend settings
    # ax.legend(
    #     loc='upper center', 
    #     bbox_to_anchor=(0.5, 1.15), 
    #     ncol=2, 
    #     fontsize=legend_sz, 
    #     frameon=False
    # )

    # Adjust layout for spacing
    plt.tight_layout()


    # Save and display the plot
    plt.savefig('lm_rs_RUI.svg')
    plt.show()


# Example call
selected_keywords = ['language modelling', 'recommendation systems']
scatter_with_fitted_curve(rows, 'publication_date', 'RUI', 'search_by_keywords', selected_keywords, remove_outliers=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from tqdm import tqdm
def gene_heatmap_plot(rows, value_col, activation_cols):
    """
    Creates an aesthetically pleasing binary heatmap with TNCSI as a gradient along the X-axis.

    Parameters:
    rows (list): A list of row objects with attributes matching column names.
    value_col (str): The name of the attribute used for sorting (e.g., TNCSI).
    activation_cols (list): A list of column names indicating binary activation (0/1).
    """
    # Filter valid rows (skip if value_col or any activation column is None)
    valid_rows = [
        row for row in rows
        if getattr(row, value_col) is not None
        and all(getattr(row, col) is not None for col in activation_cols)
    ]

    # Sort rows by the TNCSI value (from low to high)
    sorted_rows = sorted(valid_rows, key=lambda row: getattr(row, value_col))

    # Extract TNCSI values for quantile calculations
    tn_values = np.array([getattr(row, value_col) for row in sorted_rows])

    # Create a 2D matrix for the binary activation states
    activation_matrix = np.zeros((len(activation_cols), len(sorted_rows)))

    # Fill the matrix with binary values (0 or 1) from the activation columns
    for row_idx, row in enumerate(sorted_rows):
        for col_idx, col in enumerate(activation_cols):
            activation_matrix[col_idx, row_idx] = getattr(row, col)

    # Define custom colormap: Inactive (#FCF0EB), Active (#920A10)
    cmap = ListedColormap(["#EAEAF2", "steelblue"])

    # Create the heatmap
    plt.figure(figsize=(6,3))  # Adjust height dynamically
    im = plt.imshow(activation_matrix, aspect='auto', cmap=cmap, interpolation='nearest')

    # Calculate quantile positions and corresponding TNCSI values
    quantiles = np.linspace(0, 1, 5)  # 0%, 25%, 50%, 75%, 100%
    quantile_positions = (quantiles * (len(tn_values) - 1)).astype(int)  # Indices of quantiles
    quantile_labels = tn_values[quantile_positions]  # TNCSI values at these quantiles

    # Add X-axis labels for TNCSI quantiles
    plt.xticks(
        ticks=quantile_positions,
        labels=[f'{val:.2f}' for val in quantile_labels],
        fontsize=18
    )
    plt.xlabel(f'{value_col}',fontsize=18)

    # Add Y-axis labels for activation columns
    plt.yticks(
        ticks=np.arange(len(activation_cols)),
        labels=activation_cols,
        fontsize=18
    )

    # Remove gridlines
    plt.grid(False)

    # Tighten the layout for better appearance
    plt.tight_layout()

    # Set the title
    # plt.title(f'Binary Activation Heatmap across {value_col}')
    file_path = "review_feature_gene_express.svg"
    plt.savefig(file_path,dpi=300,bbox_inches='tight',)
    # Show the plot
    plt.show()
import datetime
def get_valid_rows(session):
    valid_rows = []


# 定义2022年1月1日和2023年1月1日的时间范围
    start_date = datetime.datetime(2010, 1, 1)
    end_date = datetime.datetime(2025, 1, 1)
    
    # 查询 publication_date 在 2022 年内的数据
    results = session.query(PaperMapping).filter(
        PaperMapping.publication_date >= start_date,
        PaperMapping.publication_date < end_date
    ).all()


    for row in tqdm(results):
        if row.is_review == 1 and row.valid == 1 :#and row.s2_pub_info != "arXiv.org"
            # if row.RUI and row.IEI and row.RAD and row.CDR:
            valid_rows.append(row)
    return valid_rows

rows = get_valid_rows(session)
print(len(rows))
# Example usage:
gene_heatmap_plot(rows, 'TNCSI', ['PRISMA', 'application','benchmark','preliminary','taxonomy','future',])
```

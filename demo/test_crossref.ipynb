```python
import requests

def get_publication_date(paper_title,PE='oceanytech@gmail.com'):
    # 使用Crossref API进行查询
    url = f'https://api.crossref.org/works?query={paper_title}&mailto={PE}'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        items = data['message']['items']

        if len(items) > 0:
            # 获取第一篇匹配的论文的发表日期
            publication_date = items[0]['created']['date-parts'][0]
            return publication_date

    return None

# 输入论文标题
paper_title = 'Segment Anything'

# 查询发表日期
publication_date = get_publication_date(paper_title)
print(publication_date)
if publication_date:
    # 提取年份和月份
    year, month = publication_date[:2], publication_date[2:]
    print(f'The paper "{paper_title}" was published in {month}/{year}.')
else:
    print('Publication date not found.')
```

```python
import requests

def get_citations(doi):
    # 使用Crossref API进行查询
    url = f'https://api.crossref.org/works/{doi}/references'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        citations = data['message']['items']
        return citations

    return None

# 输入文献的DOI
doi = '10.1109/access.2021.3128536'

# 查询引用文献
citations = get_citations(doi)

if citations:
    print(f'The paper with DOI "{doi}" has the following citations:')
    for citation in citations:
        title = citation.get('title', 'N/A')
        author = citation.get('author', 'N/A')
        print(f'- {title} by {author}')
else:
    print('Citations not found.')
```

```python
import requests

def get_cited_by(doi):
    # 使用 Crossref API 的 Cited-by Endpoint 进行查询
    url = f'http://api.crossref.org/works?filter=has-references:true'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        print(data)
        cited_by = data['message']['items']
        return cited_by

    return None

# 输入论文的 DOI
doi = '10.1109/access.2021.3128536'

# 获取引用了指定文献的工作列表
cited_by_list = get_cited_by(doi)

if cited_by_list:
    print(f'Works that cite the paper with DOI "{doi}":')
    for work in cited_by_list:
        title = work['title'][0] if 'title' in work else 'N/A'
        doi = work['DOI'] if 'DOI' in work else 'N/A'
        print(f'- Title: {title}')
        print(f'  DOI: {doi}')
        print('---')
else:
    print('No works found that cite the specified paper.')
```

```python
import requests

def get_crossref_articles(query, start_date, end_date):
    base_url = 'https://api.crossref.org/works'
    params = {
        'query': query,
        'filter': f'from-pub-date:{start_date},until-pub-date:{end_date}',
        'rows': 20
    }

    response = requests.get(base_url, params=params)
    data = response.json()
    print(data)
    total_results = data['message'] 
    return total_results

# 示例用法
query = 'segment anything' # 指定领域的查询词
start_date = '2023-03-01'
end_date = '2023-07-31'

article_count = get_crossref_articles(query, start_date, end_date)
print(f'Total articles in {query} from {start_date} to {end_date}: {article_count}')
```
